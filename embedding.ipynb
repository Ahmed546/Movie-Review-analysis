```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent = ['the glass of milk',
        'the glass of juice',
        'the cup of tea',
        'I am a good boy',
        'I am a good developer',
        'understand the meaning of words',
        'your videos are good']
```

```python
sent
```

```python
##Vocab size
voc_size = 10000
```

```python
# one hot representation
one_hot_repr=[one_hot(words,voc_size)for words in sent]
one_hot_repr
```

```python
## word embedding representation
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils  import pad_sequences
from tensorflow.keras.models import Sequential



```

```python
import numpy as np
```

```python
sent_length = 8
embedded_docs = pad_sequences(one_hot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
# Define the input and embedding dimensions
input_dim = 10000   # vocabulary size (adjust this to get close to 100,000 params)
output_dim = 10     # embedding output size
input_length = 8    # sequence length (None represents dynamic batch size)

# Create the Embedding layer
embedding_layer = Embedding(input_dim=input_dim, output_dim=output_dim, input_length=input_length)

# Build the model with a placeholder input
model =Sequential([
    embedding_layer
])

# Summary to check output shape and params
model.build(input_shape=(None, input_length))
model.summary()
```

```python
model.predict(embedded_docs)
```
